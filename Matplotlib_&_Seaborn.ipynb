Matplotlib

Line Plot:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]
```

create a line plot

```python
plt.plot(x,y, marker='.', markersize=8, color='purple', linestyle='-.', linewidth=2)

#create labels and title

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("X versus Y")
plt.grid(True)

plt.show()
```

```python
import pandas as pd

#creating a dataframe

data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [500, 800, 1700, 1500, 2000]
}
df = pd.DataFrame(data)

#line plot
plt.plot(df['Year'], df['Sales'], marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.grid(True)

#x-axis ticks at integer values
plt.xticks(df['Year'])

plt.show()
```

Bar Plot:

```python
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Sales': [500, 800, 1200, 1500]
}

df = pd.DataFrame(data)

#create a bar plot

plt.bar(df['Category'], df['Sales'], color='orange')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales By Category')

plt.show()
```

Scatter Plot:

```python
#creating a dataframe

data = {
    'Age': [25, 30, 22, 35, 28],
    'Income': [5000, 60000, 4500, 70000, 55000]
}
df = pd.DataFrame(data)

#line plot
plt.scatter(df['Age'], df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Income vs. Age')
plt.grid(True)

plt.show()
```

Histogram:

creating a data frame

```python
data = {
    'Scores': [85, 78, 92, 65, 80, 75, 88, 90, 82, 70]
}
df = pd.DataFrame(data)

#line plot
plt.hist(df['Scores'], width = 1)
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Scores')
plt.grid(True)

plt.show()
```

Box Plot:

```python
data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

plt.boxplot(data)
plt.show()
```
